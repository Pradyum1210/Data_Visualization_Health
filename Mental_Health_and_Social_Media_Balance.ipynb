```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/Mental_Health_and_Social_Media_Balance_Dataset.csv')
df
```

```python
df['Age_Group'] = pd.cut(df['Age'],
                         bins=[0, 20, 25, 30, 35, 40, 45, 50, 100],
                         labels=['15–20', '21–25', '26–30', '31–35', '36–40', '41–45', '46–50', '50+'])

```

```python
# Create a bar chart showing the average stress level for each age group.

stress_level = df['Stress_Level(1-10)']
age = df['Age']

plt.bar( age,stress_level, color='skyblue')
plt.title('Stress level for each age group')
plt.xlabel('Age Group')
plt.ylabel('Stress Level')
plt.show()
```

```python
# Plot a histogram to show the distribution of screen time per day across all participants.
sns.histplot(df['Daily_Screen_Time(hrs)'],kde = True)
plt.title('Screen time per day')
plt.xlabel('Daily screen time(hrs)')
plt.ylabel('People')
plt.show()
```

```python
# Draw a pie chart representing the percentage of participants by gender.

gender_count = df['Gender'].value_counts()
plt.pie(gender_count,labels = gender_count.index ,autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
# Use a box plot to compare the anxiety score across different age groups.
sns.boxplot(x = df['Age_Group'] , y = df['Stress_Level(1-10)'])
plt.xlabel('Age Group')
plt.ylabel('Stress Level')
plt.title('Anxiety score across different age group')
```

```python
# Create a scatter plot showing the relationship between screen time (hours/day) and sleep duration (hours).

sns.scatterplot(x = df['Daily_Screen_Time(hrs)'] , y = df['Sleep_Quality(1-10)'],color = 'blue')
plt.title('Screen Time vs Sleep duration')
plt.xlabel('Screen Time')
plt.ylabel('Sleep Duration')
```

```python
# Make a line chart showing the average mood score across different times of the day.

avg_happiness = df.groupby('Daily_Screen_Time(hrs)')['Happiness_Index(1-10)'].mean()

plt.plot(avg_happiness.index, avg_happiness.values, marker='o', color='orange')
plt.title('Average Happiness vs Daily Screen Time')
plt.xlabel('Daily Screen Time (hours)')
plt.ylabel('Average Happiness Index')
plt.grid(True)
plt.show()
```

```python
# Plot a heatmap showing correlations between screen time, sleep duration and stress level.

corr_relation = df[['Daily_Screen_Time(hrs)','Sleep_Quality(1-10)','Stress_Level(1-10)']]

sns.heatmap(corr_relation.corr(),annot = True)
```

```python
# Create a stacked bar chart showing mental health issues (like anxiety, depression, etc.) distribution across genders.

df.columns
plt.bar(df['Stress_Level(1-10)'],df['Age'],color = 'salmon')
plt.xlabel('Stress Level')
plt.ylabel('Age')
plt.show()
```

```python
# Draw a violin plot to show how stress level varies with social media usage.

sns.violinplot(x = df['Stress_Level(1-10)'], y = df['Daily_Screen_Time(hrs)'])
plt.xlabel('Stress Level')
plt.ylabel('Screen Time(hrs)')
```

```python
# Use a bar chart to visualize the average happiness score for each social media platform.

avg_happiness = df.groupby('Social_Media_Platform')['Happiness_Index(1-10)'].mean()

plt.bar(avg_happiness.index,avg_happiness.values)
plt.xlabel('Social medial platform')
plt.ylabel('Happiness Index (1-10)')
plt.title('Average happiness score for each social media platform')
```

```python
# Make a scatter plot with regression line between screen time and mental wellbeing score.

sns.lmplot(
    x='Daily_Screen_Time(hrs)',
    y='Stress_Level(1-10)',
    data=df,
    line_kws={'color': 'blue'}
)
plt.title('Relationship Between Screen Time and Mental Wellbeing')
plt.xlabel('Daily Screen Time (hours)')
plt.ylabel('Happiness Index (1–10)')
plt.show()
```

```python
# Create a donut chart showing percentage of people using social media for more than 4 hours/day vs less than 4 hours/day.

more_than = df[df['Daily_Screen_Time(hrs)']>4][['Daily_Screen_Time(hrs)']]
less_than = df[df['Daily_Screen_Time(hrs)']<4][['Daily_Screen_Time(hrs)']]



labels = ['More than 4 hrs', 'Less than 4 hrs']
values = [len(more_than), len(less_than)]


plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99'])


centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Screen Time Comparison')
plt.show()
```

```python
# Plot a countplot of participants based on their preferred social media platform.
# df.columns
sns.countplot(df['Social_Media_Platform'], palette='pastel')
plt.title('Count of Participants by preferred social media platform')
plt.ylabel('Platform')
plt.xlabel('Count')
plt.show()
```

```python
# Create a grouped bar chart showing average sleep duration by gender and screen time category.

avg_sleep_gender = df.groupby('Gender')['Sleep_Quality(1-10)'].mean()
avg_sleep_screen = df.groupby('Daily_Screen_Time(hrs)')['Sleep_Quality(1-10)'].mean()

plt.subplot(1,2,1)
plt.bar(avg_sleep_gender.index,avg_sleep_gender.values)
plt.title("Avg sleep by gender")
plt.xlabel('Gender')
plt.ylabel('Average sleep')

plt.subplot(1,2,2)
plt.bar(avg_sleep_screen.index,avg_sleep_screen.values)
plt.title("Avg sleep by screen time")
plt.xlabel('Screen time')
plt.ylabel('Average sleep')

plt.show()
```

```python
# Draw a pairplot to visualize relationships among stress, anxiety, sleep, and screen time.

data = df[['Stress_Level(1-10)','Sleep_Quality(1-10)','Daily_Screen_Time(hrs)']]

sns.pairplot(data)
```

```python
# Plot a bar chart showing the average productivity score for people with different levels of social media addiction.

df['Addiction_Level'] = pd.cut(df['Daily_Screen_Time(hrs)'],
                               bins = [0,2,4,6,10],
                               labels = ['Low','Moderate','High','Very High'])

data = df.groupby('Addiction_Level')['Sleep_Quality(1-10)'].mean()

plt.bar(data.index,data.values,color = 'brown')
plt.title('Social media addiction vs Sleep Quality')
plt.xlabel('Social media addiction')
plt.ylabel('Sleep Quality(1-10)')
```

```python
# Make a histogram to show how daily screen time is distributed among people aged 18–25.


data = df.loc[(df['Age'] >= 18) & (df['Age'] <= 25)][['Age', 'Daily_Screen_Time(hrs)']]



plt.hist(data['Daily_Screen_Time(hrs)'], bins=9,color='skyblue', edgecolor='black')


plt.title('Distribution of Daily Screen Time (Age 18–25)')
plt.xlabel('Daily Screen Time (hours)')
plt.ylabel('Number of People')

plt.show()

```

```python
# Use a heatmap to show average anxiety score by age group and gender.
df['Gender_Int'] = df['Gender'].map({'Male': 1, 'Female': 0,'Other':3})

relation = df[['Age','Stress_Level(1-10)','Gender_Int']]
relation.corr()

sns.heatmap(relation.corr(),annot = True)
```

```python
# Create a line chart comparing average stress level across different days of the week.


avg_stress = df.groupby('Exercise_Frequency(week)')['Stress_Level(1-10)'].mean()


plt.plot(avg_stress.index, avg_stress.values, marker='o')
plt.title('Average Stress Level Across Days of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Stress Level')
plt.grid(True)
plt.show()


```

```python
# Make a bubble chart where bubble size represents sleep hours and axes show screen time vs anxiety score.

plt.figure(figsize=(8,6))

plt.scatter(
    df['Daily_Screen_Time(hrs)'],
    df['Stress_Level(1-10)'],
    s = df['Sleep_Quality(1-10)'] * 50,
    alpha = 0.6,
    color = 'skyblue',
    edgecolors='black'
)

plt.title('Bubble Chart: Screen Time vs Stress Level (Bubble = Sleep Quality)')
plt.xlabel('Daily Screen Time (hrs)')
plt.ylabel('Stress Level (1–10)')

plt.show()
```

```python
# Plot a horizontal bar chart showing top 5 factors affecting mental health (if a column for factors exists).


factors = ['Daily_Screen_Time(hrs)', 'Sleep_Quality(1-10)',
           'Days_Without_Social_Media', 'Exercise_Frequency(week)',
           'Happiness_Index(1-10)']


correlation = df[factors].corrwith(df['Stress_Level(1-10)']).abs()


top_5 = correlation.sort_values(ascending=False).head(5)


plt.figure(figsize=(8,5))
plt.barh(top_5.index, top_5.values, color='orange')

plt.title("Top 5 Factors Affecting Mental Health (Based on Correlation)")
plt.xlabel("Correlation with Stress Level")
plt.ylabel("Factors")

plt.show()
```

```python
# Create a pie chart to show the ratio of participants reporting depression vs no depression.

df['Depression'] = df['Stress_Level(1-10)'].apply(lambda x: 'Depression' if x>=7 else 'No Depression')

depression_count = df['Depression'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    depression_count.values,
    labels = depression_count.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['red','skyblue']
)

plt.title('Depression vs No Depression Ratio')
plt.show()
```

```python
# Draw a bar chart comparing average happiness levels among people using different social media platforms.

avg_happiness = df.groupby('Social_Media_Platform')['Happiness_Index(1-10)'].mean()


plt.figure(figsize=(8,5))
plt.bar(avg_happiness.index, avg_happiness.values, color='skyblue', edgecolor='black')

plt.title('Average Happiness Levels by Social Media Platform')
plt.xlabel('Social Media Platform')
plt.ylabel('Average Happiness Score')

plt.xticks(rotation=45)
plt.show()
```

```python
# Plot a scatter plot between screen time and happiness score, colored by age group.
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Daily_Screen_Time(hrs)',
    y='Happiness_Index(1-10)',
    hue='Age_Group',
    palette='viridis',
    s=80
)

plt.title('Screen Time vs Happiness Score (Colored by Age Group)')
plt.xlabel('Daily Screen Time (hrs)')
plt.ylabel('Happiness Score (1–10)')
plt.legend(title='Age Group')
plt.show()
```

```python
# Plot a countplot for frequency of social media use (e.g., rarely, sometimes, often, always).
df['Usage_Frequency'] = pd.cut(
    df['Daily_Screen_Time(hrs)'],
    bins=[0, 2, 4, 6, 24],
    labels=['Rarely', 'Sometimes', 'Often', 'Always']
)
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='Usage_Frequency', palette='pastel')

plt.title('Frequency of Social Media Use')
plt.xlabel('Usage Category')
plt.ylabel('Number of Participants')

plt.show()
```

```python
# Make a box plot comparing sleep duration across levels of social media usage.
plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x='Usage_Frequency',
    y='Sleep_Quality(1-10)',
    palette='Set2'
)

plt.title('Sleep Quality Across Different Levels of Social Media Usage')
plt.xlabel('Social Media Usage Frequency')
plt.ylabel('Sleep Quality (1–10)')
plt.show()
```
